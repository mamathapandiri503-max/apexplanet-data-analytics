```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Quantity'], bins=30)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Quantity'], bins=30)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
print(df.columns)
```

```python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Quantity'], bins=30)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['TotalSales'], bins=30)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Quantity'])
plt.title('Boxplot of Quantity')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot of Unit Price')
plt.show()
```

```python
country_sales = df.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(10)

country_sales.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Countries by Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(12,5))
plt.title('Top 10 Best Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=90)
plt.show()
```

```python
corr = df[['Quantity', 'UnitPrice', 'CustomerID', 'TotalSales']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['TotalSales'].sum()
```

```python
monthly_sales.plot(figsize=(12,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

# Key Insights
1. The United Kingdom generated the highest sales among all countries.
2. Most customer orders contained small quantities of products.
3. A few products contributed significantly to the overall sales.
4. Monthly sales varied over time, indicating seasonal trends.
5. Quantity sold and Total Sales showed a positive relationship.
